```python
import lightgbm as lgb
import pandas as pd
import numpy as np
from scipy import sparse
from sklearn.feature_selection import SelectFromModel

select_tr_x = sparse.load_npz('data/tr_x_3_6482.npz')
select_val_x = sparse.load_npz('data/val_x_3_6482.npz')
select_test_x = sparse.load_npz('data/test_x_3_6482.npz')

tr_y = pd.read_csv('data/tr_y.csv')
val_y = pd.read_csv('data/val_y.csv')
res = pd.read_csv('data/res.csv')
```

```python
clf = lgb.LGBMClassifier(
    boosting_type='gbdt', num_leaves=600, reg_alpha=0.0, reg_lambda=1,
    max_depth=-1, n_estimators=5000, objective='binary',
    subsample=0.7, colsample_bytree=0.7, subsample_freq=1,
    learning_rate=0.03, min_child_weight=10, device='gpu', random_state=2018
)
clf.fit(select_tr_x, tr_y, eval_set=[(select_val_x, val_y)], eval_metric='auc', early_stopping_rounds=100)
print('training completed!')
```

```python
print('reading test...')

print('test shape:', select_test_x.shape)
print('predicting...')
res['score'] = clf.predict_proba(select_test_x)[:, 1]
res['score'] = res['score'].apply(lambda x: float('%.6f' % x))
print('saving submission...')
res.to_csv('data_0511/result/submission.csv', index=False)
print('completed!')
```

```python
clf.booster_.save_model('data/result/05111200_ml600_mcw10_sa07_cs07_1249_749547_important_feature3')
```

```python
# >1
clf.feature_importances_[clf.feature_importances_>1].size
```

```python
# >=3
selection = SelectFromModel(clf, threshold=3, prefit=True)
select_tr_x = selection.transform(select_tr_x)
select_val_x = selection.transform(select_val_x)
select_test_x = selection.transform(select_test_x)


sparse.save_npz('data/tr_x_3_6482.npz', select_tr_x)
sparse.save_npz('data/val_x_3_6482.npz', select_val_x)
sparse.save_npz('data/test_x_3_6482.npz', select_test_x)
```
