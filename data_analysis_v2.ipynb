### 由以下可知，各属性特征种类可按词向量的方式处理

```python
print('advertiserId number: ', ad_feature['advertiserId'].drop_duplicates().count())
print('campaignId number: ', ad_feature['campaignId'].drop_duplicates().count())
```

```python
advertiserId = ad_feature['advertiserId']
advertiserId = np.log(advertiserId)
advertiserId.describe()
```

```python
ad_feature['advertiserId'] = advertiserId
```

```python
ad_feature['campaignId'].describe()
```

```python
campaignId = ad_feature['campaignId']
campaignId = np.log(campaignId)
campaignId.describe()
```

```python
ad_feature['campaignId'] = campaignId
```

```python
print('creativeId number: ', ad_feature['creativeId'].drop_duplicates().count())
print('creativeSize number: ', ad_feature['creativeSize'].drop_duplicates().count())
```

```python
creativeId = ad_feature['creativeId']
creativeId = np.log(creativeId)
creativeId.describe()
```

```python
ad_feature['creativeId'] = creativeId
```

```python
ad_feature['creativeSize'].describe()
```

```python
adCategoryId = ad_feature['adCategoryId']
adCategoryId = np.log(adCategoryId)
adCategoryId.describe()
```

```python
ad_feature['adCategoryId'] = adCategoryId
```

```python
productId = ad_feature['productId']
productId = productId + 1
productId = np.log(productId)
productId.describe()
```

```python
ad_feature['productId'] = productId
```

```python
ad_feature['productType'].describe()
```

```python
print('adCategoryId number: ', ad_feature['adCategoryId'].drop_duplicates().count())
print('productId number: ', ad_feature['productId'].drop_duplicates().count())
print('productType number: ', ad_feature['productType'].drop_duplicates().count())
```

```python
ad_feature.describe()
```

```python
ad_norm = ad_feature.copy()
ad_norm['advertiserId'] = (ad_norm['advertiserId'] - ad_norm['advertiserId'].min()) / (ad_norm['advertiserId'].max() - ad_norm['advertiserId'].min())
ad_norm['campaignId'] = (ad_norm['campaignId'] - ad_norm['campaignId'].min()) / (ad_norm['campaignId'].max() - ad_norm['campaignId'].min())
ad_norm['creativeId'] = (ad_norm['creativeId'] - ad_norm['creativeId'].min()) / (ad_norm['creativeId'].max() - ad_norm['creativeId'].min())
ad_norm['creativeSize'] = (ad_norm['creativeSize'] - ad_norm['creativeSize'].min()) / (ad_norm['creativeSize'].max() - ad_norm['creativeSize'].min())
ad_norm['adCategoryId'] = (ad_norm['adCategoryId'] - ad_norm['adCategoryId'].min()) / (ad_norm['adCategoryId'].max() - ad_norm['adCategoryId'].min())
ad_norm['productId'] = (ad_norm['productId'] - ad_norm['productId'].min()) / (ad_norm['productId'].max() - ad_norm['productId'].min())
ad_norm['productType'] = (ad_norm['productType'] - ad_norm['productType'].min()) / (ad_norm['productType'].max() - ad_norm['productType'].min())
ad_norm.describe()
```

```python
train_train = pd.read_csv('data/train_train.csv')
train_valid = pd.read_csv('data/train_valid.csv')
```

```python
train_train_dropped = train_train.drop(columns=['advertiserId', 'campaignId', 'creativeId', 'creativeSize', 'adCategoryId', 'productId', 'productType'])
train_train_dropped.head()
```

```python
train_valid_dropped = train_valid.drop(columns=['advertiserId', 'campaignId', 'creativeId', 'creativeSize', 'adCategoryId', 'productId', 'productType'])
```

```python
train_train_v2 = pd.merge(train_train_dropped, ad_norm, on='aid')
train_valid_v2 = pd.merge(train_valid_dropped, ad_norm, on='aid')
```

```python
train_train_v2['advertiserId'].describe()
```

```python
from sklearn.utils import shuffle
```

```python
train_train_v2 = shuffle(train_train_v2)
train_valid_v2 = shuffle(train_valid_v2)
```

```python
train_train_v2.to_csv('data/train_train_v2.csv', columns=train_train.columns, index=False)
train_valid_v2.to_csv('data/train_valid_v2.csv', columns=train_train.columns, index=False)
```
