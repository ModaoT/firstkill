```python
import pandas as pd
```

```python
ad_feature_mapped = pd.read_csv('data/ad_feature_mapped.csv')
user_feature_base_normalized = pd.read_csv('data/user_feature_base_normalized.csv')
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
print('user feature number: ', user_feature_base_normalized['uid'].drop_duplicates().count())
print('train uid number: ', train['uid'].drop_duplicates().count())
print('test uid number:  ', test['uid'].drop_duplicates().count())
print('test + train:  ', test['uid'].drop_duplicates().count() + train['uid'].drop_duplicates().count())
```

```python
print('ad feature number: ', ad_feature_mapped['aid'].drop_duplicates().count())
print('train aid number: ', train['aid'].drop_duplicates().count())
print('test aid number:  ', test['aid'].drop_duplicates().count())
```
